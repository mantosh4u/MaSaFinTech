```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ROE = P/B/P/E
# P/B = ROE*P/E
# E/P = ROE/P/B
# P/B is directly proportional to the ROE for constant P/E
# E/P is directly proportional to the ROE for constant P/B
```

```python
x = np.linspace(5,40,20)
y = 10*x/100
plt.plot(x, y, '-g', label='P/B & ROE with P/E = 10')
y = 20*x/100
plt.plot(x, y, '-r', label='P/B & ROE with P/E = 20')
y = 5*x/100
plt.plot(x, y, '-y', label='P/B & ROE with P/E = 5')
plt.title('Relation Between P/B & ROE', color ='blue')
plt.xlabel('ROE', color='blue')
plt.ylabel('P/B', color='blue')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
x = np.linspace(5,40,20)
y = 1*100/x
plt.plot(x, y, '-g', label='P/E = ROE with P/B = 1')
y = 2*100/x
plt.plot(x, y, '-r', label='P/E = ROE with P/B = 2')
y = 0.5*100/x
plt.plot(x, y, '-y', label='P/E = ROE with P/B = 0.5')
plt.title('Relation Between P/E & ROE', color ='blue')
plt.xlabel('ROE', color='blue')
plt.ylabel('P/E', color='blue')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
