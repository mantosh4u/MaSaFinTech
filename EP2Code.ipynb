```python
import pandas as pd
import random
```

```python
tvColumList = ["Year", "Equity", "Profit", "ROE"]


tvRoeReturn    = 10   # ROE value which would be constant over the time
tvIntialEquity = 10   # initial captial(equity)
tvNumberOfYear = 16   # let's us calculate it for 15 year

tvYearList   = []
tvEquityList = []
tvProfitList = []
tvRoeList    = []

tvYearList.append(0)
tvEquityList.append(tvIntialEquity)
tvProfitList.append(tvIntialEquity*tvRoeReturn/100)
tvRoeList.append(tvRoeReturn)

# To make it more realistic to the real world companies, which could earn
# higher ROE in the good time and medicore ROE in bad time of business, let
# us use the random number generator instead of constant number.
for i in range(1,tvNumberOfYear):
    n = random.randint(10,20)
    tvRoeList.append(n)
```

```python
for index in range(1, tvNumberOfYear):
    tvYearList.append(index)
    tvTotalCurrentYearEquity = tvEquityList[index -1] + tvProfitList[index - 1]
    tvEquityList.append(tvTotalCurrentYearEquity)
    tvRoeReturn = tvRoeList[index]
    tvProfitList.append(tvTotalCurrentYearEquity*tvRoeReturn/100)

    

tvEquityList = [round(item,1) for item in tvEquityList]
tvProfitList = [round(item,1) for item in tvProfitList]


inputDic = {tvColumList[0]:tvYearList,\
            tvColumList[1]:tvEquityList,\
            tvColumList[2]:tvProfitList,\
            tvColumList[3]:tvRoeList
            }

df = pd.DataFrame(inputDic, columns=tvColumList)
df = df.reset_index(drop=True)
df = df.set_index(tvColumList[0])
```

```python
# let us see the content of the dataframe at this point of time.
df.head(30)
```

```python
# Now lets us calculate the series of values of P/E when MCap is equal
# to the underlying business value of the company.
tvMCapValueList  =  []
tvPEValueList    =  []


for index in range(0, tvNumberOfYear):
    tvMCap = 1*tvEquityList[index]
    tvMCap = round(tvMCap,1)
    tvMCapValueList.append(tvMCap)
    tvProfit = tvProfitList[index]
    tvPE = tvMCap/tvProfit
    tvPE = round(tvPE,1)
    tvPEValueList.append(tvPE)
```

```python
# Now lets us create a new dataframe with all five member together.
tvFinalColumList = ["MCap=Value", "P/E","Year", "Equity", "Profit", "ROE"]

inputDic = {tvFinalColumList[0]:tvMCapValueList,\
            tvFinalColumList[1]:tvPEValueList,\
            tvFinalColumList[2]:tvYearList,\
            tvFinalColumList[3]:tvEquityList,\
            tvFinalColumList[4]:tvProfitList,\
            tvFinalColumList[5]:tvRoeList
            }

dfFinal = pd.DataFrame(inputDic, columns=tvFinalColumList)
dfFinal = dfFinal.reset_index(drop=True)
dfFinal = dfFinal.set_index(tvColumList[0])
```
